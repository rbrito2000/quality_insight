```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Carregar o dataset
df = pd.read_csv('../data/weaving_rejection_dataset_updated.csv')
```

```python
# Calcular a matriz de correlação
correlation_matrix = df.corr()
```

```python
# Somar as correlações de cada linha
correlation_sums = correlation_matrix.sum(axis=1)

# Ordenar as somas por ordem decrescente
correlation_sums_sorted = correlation_sums.sort_values(ascending=False)
```

```python
# Imprimir o resultado
print("Somas das correlações por linha (ordenadas):")
print(correlation_sums_sorted)
```

```python
# Verificar os nomes das colunas
print(df.columns)
```

```python
# Pares de variáveis para análise
pair_variables = [
    ['Req_Finish_Fabrics', 'Req_grey_fabric'],
    ['Req_Finish_Fabrics', 'Req_beam_length.yds.'],
    ['Shrink_allow', 'Fabric_Allowance'],
    ['Req_beam_length.yds.', 'Total_Pdn.yds.'],
    ['Req_beam_length.yds.', 'Req_grey_fabric'],
    ['weft_count', 'warp_count']
]

# Função para gerar pairplots individualmente
def plot_pair(var1, var2, hue_var='Rejection'):
    sns.lmplot(x=var1, y=var2, hue=hue_var, data=df, palette='Set2', fit_reg=False, markers='o')  # Um único marcador
    plt.title(f'Pairplot de {var1} e {var2} com a influência de {hue_var}')
    plt.show()

# Gerar gráficos individuais
for pair in pair_variables:
    plot_pair(pair[0], pair[1])
```
